### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Ingresos por tienda
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

# Mostrar resultados
print(f'💰 Ingreso total Tienda 1: ${ingreso_tienda1:,.2f}')
print(f'💰 Ingreso total Tienda 2: ${ingreso_tienda2:,.2f}')
print(f'💰 Ingreso total Tienda 3: ${ingreso_tienda3:,.2f}')
print(f'💰 Ingreso total Tienda 4: ${ingreso_tienda4:,.2f}')
```

# 2. Ventas por categoría

```python
import pandas as pd
```

```python
# Cantidad de productos vendidos por categoría en cada tienda
cat_t1 = tienda['Categoría del Producto'].value_counts()
cat_t2 = tienda2['Categoría del Producto'].value_counts()
cat_t3 = tienda3['Categoría del Producto'].value_counts()
cat_t4 = tienda4['Categoría del Producto'].value_counts()
```

```python
print("📦 Productos vendidos por categoría - Tienda 1:")
print(cat_t1)
print("\n📦 Productos vendidos por categoría - Tienda 2:")
print(cat_t2)
print("\n📦 Productos vendidos por categoría - Tienda 3:")
print(cat_t3)
print("\n📦 Productos vendidos por categoría - Tienda 4:")
print(cat_t4)
```

# 3. Calificación promedio de la tienda


```python
# Calificación promedio de cada tienda
promedio_t1 = tienda['Calificación'].mean()
promedio_t2 = tienda2['Calificación'].mean()
promedio_t3 = tienda3['Calificación'].mean()
promedio_t4 = tienda4['Calificación'].mean()
```

```python
print(f'⭐ Calificación promedio Tienda 1: {promedio_t1:.2f}')
print(f'⭐ Calificación promedio Tienda 2: {promedio_t2:.2f}')
print(f'⭐ Calificación promedio Tienda 3: {promedio_t3:.2f}')
print(f'⭐ Calificación promedio Tienda 4: {promedio_t4:.2f}')
```

# 4. Productos más y menos vendidos

```python
# Productos más y menos vendidos - Tienda 1
productos_t1 = tienda['Producto'].value_counts()
mas_vendido_t1 = productos_t1.idxmax()
menos_vendido_t1 = productos_t1.idxmin()

# Tienda 2
productos_t2 = tienda2['Producto'].value_counts()
mas_vendido_t2 = productos_t2.idxmax()
menos_vendido_t2 = productos_t2.idxmin()

# Tienda 3
productos_t3 = tienda3['Producto'].value_counts()
mas_vendido_t3 = productos_t3.idxmax()
menos_vendido_t3 = productos_t3.idxmin()

# Tienda 4
productos_t4 = tienda4['Producto'].value_counts()
mas_vendido_t4 = productos_t4.idxmax()
menos_vendido_t4 = productos_t4.idxmin()
```

```python
print("🏆 Productos más vendidos:")
print(f"Tienda 1: {mas_vendido_t1}")
print(f"Tienda 2: {mas_vendido_t2}")
print(f"Tienda 3: {mas_vendido_t3}")
print(f"Tienda 4: {mas_vendido_t4}")

print("\n📉 Productos menos vendidos:")
print(f"Tienda 1: {menos_vendido_t1}")
print(f"Tienda 2: {menos_vendido_t2}")
print(f"Tienda 3: {menos_vendido_t3}")
print(f"Tienda 4: {menos_vendido_t4}")
```

# 1. Gráfico circular – Participación de Cada Tienda en los Ingresos Totales

```python
import matplotlib.pyplot as plt

# Ingresos por tienda (reemplazá con tus valores si ya los calculaste)
ingresos = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Gráfico de torta
plt.figure(figsize=(8, 6))
plt.pie(
    ingresos,
    labels=tiendas,
    autopct='%1.1f%%',
    startangle=90,
    colors=['#3CB371', '#6495ED', '#F08080', '#FFD700'],
    shadow=True
)

plt.title('💰 Participación de Cada Tienda en los Ingresos Totales')
plt.axis('equal')  # Para que el gráfico sea circular
plt.show()
```

#2. Gráfico Barra– Calificación promedio de las tiendas

```python
# Datos
calificaciones = [3.98, 4.04, 4.05, 4.00]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
colores = ['mediumseagreen', 'cornflowerblue', 'lightcoral', 'gold']

# Crear el gráfico de barras
plt.figure(figsize=(8, 5))
bars = plt.bar(tiendas, calificaciones, color=colores)

# Agregar los valores encima de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01, f'{yval:.2f}', ha='center', va='bottom')

# Título y etiquetas
plt.title('Calificaciones Promedio por Tienda')
plt.ylabel('Calificación Promedio')
plt.ylim(3.9, 4.1)  # Ajusta el eje Y para que se vea mejor el detalle
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
